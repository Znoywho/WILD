# The Pandas DataFrame Object

```python
area_dict = {
    'California': 423967,
    'Texas': 695662,
    'New York': 141297,
    'Florida': 170312, 
    'Illinois': 149995
}
area = pd.Series(area_dict)
area
```

```python
states = pd.DataFrame(
    {'population':population, # DICT
      'area': area}
)
states
```

```python
print("INDEX: ",end='')
print(states.index)
print("COlUMNS: ",end='')
print(states.columns)
```

```python
states['area']
```

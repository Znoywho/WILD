##### *CONCATENATE*

````python
import numpy as np
x = np.array([1,2,3])
x.reshape((3,1))
````

````python
import numpy as np
grid = np.arange(1,10).reshape((3,3))
print(grid)
````

````python
import numpy as np
x1 = np.arange(1,10).reshape((3,3))
x2 = np.arange(11,20).reshape((3,3))


total = np.concatenate([x1,x2]).reshape((3,6))
print(total)
````

````python
import numpy as np
np.random.seed(0) # AVOID THE DUPLICATION OF SCENARIO
x1 = np.random.randint(100,size = (10))
x2 = np.random.randint(100,size = (10))


np.vstack([x1,x2]) # vertical concatenation
````

````python
import numpy as np
np.random.seed(0)
x1 = np.random.randint(100,size = (10))
x2 = np.random.randint(100,size = (10))

np.hstack([x1,x2]) # horizontal stack
````

##### *SPILTTING*

````python
x = np.random.randint(100,size = (8))
x1, x2, x3 = np.split(x,[3,5])
print(x1,x2,x3)
````

> split N -> N + 1 subarrays

````python
grid = np.arange(16).reshape((4,4))
upper, lower = np.vsplit(grid,[2]) # SPLIT vertcally opposite to horizontally
print("UPPER:")
print(upper)
print("LOWER")
print(lower)
````

````python
import numpy as np
np.random.seed(0)

def compute_recporals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/ values[i]
    return output

values = np.random.randint(1,10,size = 5)
compute_recporals(values)
````

````python
import numpy as np
np.random.seed(0)

def compute_recporals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/ values[i]
    return output

values = np.random.randint(1,10,size = 5)
%timeit compute_recporals(values)
````

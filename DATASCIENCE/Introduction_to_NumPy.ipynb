````python
import numpy as np
np.random.seed(0)

def compute_recporals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/ values[i]
    return output

values = np.random.randint(1,10,size = 5)
compute_recporals(values)
````

````python
import numpy as np
np.random.seed(0)

def compute_recporals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/ values[i]
    return output

values = np.random.randint(1,10,size = 5)
%timeit compute_recporals(values)
````

````python
import numpy as np
alpha = np.linspace(0,np.pi,20)
print(alpha)
````

### Array Indexing: Accessing Single Elements

#### One-dimension

````python
import numpy as np
x1 = np.array([5,6,6,44])
print(x1[1])
````

#### Mutiple-dimensioon

````python
import numpy as np
x3 = np.random.randint(10,size = (2,3,4))
print(x3)
print(x3[1,2,3])
````
